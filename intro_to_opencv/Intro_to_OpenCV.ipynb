```python
import cv2

def resizeFrame(frame, f=0.5):
        #1 chieu cao, 0 chieu rong
    width = int(frame.shape[1]*f)
    height = int(frame.shape[0]*f)
    size = (width, height)
    return cv2.resize(frame, size, interpolation=cv2.INTER_AREA)

cap = cv2.VideoCapture(r"assets\720p.mp4")
while True:
    isTrue, frame = cap.read()
    resized_frame = resizeFrame(frame)
    #show video
    cv2.imshow("resized_frame",resized_frame)
    cv2.imshow("frame",frame)

    #delay 20ms, hoac dung phim bam
    if cv2.waitKey(20) & 0xFF == ord("q"):
        break

#giai phong tai nguyen
cap.release()
#dong tat ca cua so hien thi
cv2.destroyAllWindows()
```

## 3.2 V hnh v chn Text


Trong ng cnh ca OpenCV, "bng v" khng phi l mt i tng c th, m thc t l mt cch biu din hnh nh, c th hin thng qua mt mng NumPy trong Python. Di y l mt s im quan trng lin quan n vic s dng bng v (mng NumPy) trong OpenCV:




1.   **Biu Din Hnh nh:**

*   Bng v l mt mng hai chiu trong NumPy, i din cho hnh nh. Trong mng ny, mi phn t l mt pixel.
*   i vi hnh nh mu, mi pixel thng c ba gi tr (3 knh mu: B, G, R), trong khi hnh nh en trng c mt gi tr  sng.

### C Th Bn Cha bit ?

Pixel (vit tt ca picture element) l n v c bn nht ca mt hnh nh s v thnh phn c bn trong qu trnh biu din v x l hnh nh s.

Mt pixel thng bao gm cc thng tin nh sau:

*   **Mu Sc**: Trong hnh nh mu, mi pixel s c ba gi tr tng ng vi mc   (Red), xanh l cy (Green), v xanh dng (Blue). Mi gi tr ny thng nm trong khong t 0 n 255.
*   ** Sng**: Trong hnh nh en trng (grayscale), mi pixel ch cha mt gi tr  sng. Gi tr ny thng cng nm trong khong t 0 n 255, trong  0 thng l mu en v 255 l mu trng.
*   **V Tr**: Pixel cng cha thng tin v v tr ca n trong bng v hoc hnh nh.



2.   **Kch Thc Mng:**

*   Kch thc ca mng NumPy phn nh  phn gii ca hnh nh. V d, mt hnh nh c kch thc 300x400 s c mt mng NumPy kch thc (300, 400, 3) nu l hnh nh mu (3 knh mu) hoc (300, 400) nu l hnh nh en trng.

3.   **Thao Tc Vi Pixel:**

*   Bn c th trc tip truy cp v thay i gi tr ca tng pixel trong mng NumPy. iu ny rt hu ch khi bn mun thay i ni dung ca hnh nh.

4.   **Mt S Hm V v Chn Vn Bn:**

*   cv2.line(): V ng thng.
*   cv2.rectangle(): V hnh ch nht.
*   cv2.circle(): V hnh trn.
*   cv2.putText(): Chn vn bn.

5.   **Hiu Sut Cao:**

*   Mng NumPy trong OpenCV c ti u ha  hiu sut cao trong qu trnh x l hnh nh. iu ny lm cho OpenCV tr thnh mt th vin mnh m cho cc ng dng th gic my tnh v x l hnh nh trong thi gian thc.



1.   cv2.line(img, pt1, pt2, color[, thickness[, lineType[, shift]]]):

* V ng thng gia hai im pt1 v pt2 trn hnh nh img.
*   color: Mu ca ng thng.
*   thickness:  dy ca ng thng. Nu l m, ng thng s c v bng mu y .
*   lineType: Loi ng
*   shift: S bit  dch phi.

V D:

```python
import cv2
import numpy as np

# To mt hnh nh trng (mt bng v)
img = np.zeros((300, 400, 3), dtype="uint8")

# im u (pt1) v im cui (pt2) ca ng thng
pt1 = (50, 150)
pt2 = (300, 150)

# Mu ca ng thng (xanh l cy)
color = (0, 255, 0)

#  dy ca ng thng
thickness = 10

# Loi ng (cv2.LINE_AA cho ng lin vi ng gc c lm mn)
lineType = cv2.LINE_8

# V ng thng ln hnh nh
cv2.line(img, pt1, pt2, color, thickness, lineType)

# Hin th hnh nh cha ng thng
cv2.imshow("picture", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```


2.   cv2.rectangle(img, pt1, pt2, color[, thickness[, lineType[, shift]]]):


*   pt1: Ta  ca gc tri trn
*   pt2: Ta  ca gc phi di
*   color, thickness, lineType ging bn trn

V D:

```python
import cv2
import numpy as np

# To hnh nh trng (nn)
height, width = 300, 400
img = np.ones((height, width, 3), dtype=np.uint8) * 255

# c im ca hnh ch nht
pt1 = (50,50)  # Ta  ca gc tri trn
pt2 = (200, 200)  # Ta  ca gc phi di
color = (0, 0, 255)  # Mu ca hnh ch nht (mu )
thickness = 2  #  dy ca ng vin (nu l m, hnh ch nht s c v bng mu y )

# V hnh ch nht
cv2.rectangle(img, pt1, pt2, color, thickness)

# Hin th hnh nh
cv2.imshow("picture", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```



3.   cv2.circle(img, center, radius, color[, thickness[, lineType[, shift]]]):

*   center l ta  ca tm ca hnh trn.
* radius l bn knh ca hnh trn.
* color l mu ca hnh trn.
* thickness l  dy ca ng vin ca hnh trn.







```python
import cv2
import numpy as np

# To hnh nh trng (nn)
height, width = 300, 400
img = np.ones((height, width, 3), dtype=np.uint8) * 255

# c im ca hnh trn
center = (150, 150)  # Ta  tm ca hnh trn
radius = 50  # Bn knh ca hnh trn
color = (0, 255, 0)  # Mu ca hnh trn (mu xanh l cy)
thickness = 2  #  dy ca ng vin (nu l m, hnh trn s c v bng mu y )

# V hnh trn
cv2.circle(img, center, radius, color, thickness)

# Hin th hnh nh
cv2.imshow("picture", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

4. cv2.putText(img, text, org, fontFace, fontScale, color[, thickness[, lineType[shift]]])

*   img: Hnh nh ch.
* text: Chui vn bn cn chn.
* org: Ta  (x, y) ca im bt u vn bn.
* fontFace: Kiu font
* fontScale: T l kch thc ca font.
* color: Mu ca vn bn.
* thickness :  dy ca ng vin vn bn.
* lineType : Loi ng.
* shift : Nu l True, org c coi l gc tri di ca vn bn. Nu l False (mc nh), org c coi l gc tri trn ca vn bn.




```python
import cv2
import numpy as np

# To hnh nh trng (nn)
height, width = 300, 400
img = np.ones((height, width, 3), dtype=np.uint8) * 255

# Vn bn cn chn
text_to_insert = "Hello, OpenCV!"

# Ta  v font
org = (50, 150)
font_face = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1
text_color = (0, 0, 255)  # Mu 

# Chn vn bn vo hnh nh
cv2.putText(img, text_to_insert, org, font_face, font_scale, text_color, 2, cv2.LINE_AA)

# Hin th hnh nh
cv2.imshow("picture", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Face detection (Bonus)

```python
image = resizeFrame(cv2.imread(r'assets\A_Tien2.jpg'), 0.25)

grey = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)

face_cascade = cv2.CascadeClassifier(r'assets/haarcascade_frontalface_default.xml')

faces = face_cascade.detectMultiScale(grey, 1.3, 5)

for (x,y,w,h) in faces:
     cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),2)

cv2.imshow('img',cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

cv2.waitKey(0)
```

```python
import cv2
face_cascade = cv2.CascadeClassifier(r'assets/haarcascade_frontalface_default.xml')

# Create a VideoCapture object to capture video from the camera
cap = cv2.VideoCapture(0)

while True:
    # Read the frame from the camera
    ret, frame = cap.read()

    grey = cv2.cvtColor(frame , cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(grey, 1.3, 5)

    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)

    cv2.imshow('img',cv2.cvtColor(frame, 1))

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the VideoCapture object and close the window
cap.release()
cv2.destroyAllWindows()
```
