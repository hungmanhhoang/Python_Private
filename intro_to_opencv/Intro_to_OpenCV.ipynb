```python
import cv2

def resizeFrame(frame, f=0.5):
        #1 chieu cao, 0 chieu rong
    width = int(frame.shape[1]*f)
    height = int(frame.shape[0]*f)
    size = (width, height)
    return cv2.resize(frame, size, interpolation=cv2.INTER_AREA)

cap = cv2.VideoCapture(r"assets\720p.mp4")
while True:
    isTrue, frame = cap.read()
    resized_frame = resizeFrame(frame)
    #show video
    cv2.imshow("resized_frame",resized_frame)
    cv2.imshow("frame",frame)

    #delay 20ms, hoac dung phim bam
    if cv2.waitKey(20) & 0xFF == ord("q"):
        break

#giai phong tai nguyen
cap.release()
#dong tat ca cua so hien thi
cv2.destroyAllWindows()
```

## 3.2 Vẽ hình và chèn Text


Trong ngữ cảnh của OpenCV, "bảng vẽ" không phải là một đối tượng cụ thể, mà thực tế là một cách biểu diễn hình ảnh, được thể hiện thông qua một mảng NumPy trong Python. Dưới đây là một số điểm quan trọng liên quan đến việc sử dụng bảng vẽ (mảng NumPy) trong OpenCV:




1.   **Biểu Diễn Hình Ảnh:**

*   Bảng vẽ là một mảng hai chiều trong NumPy, đại diện cho hình ảnh. Trong mảng này, mỗi phần tử là một pixel.
*   Đối với hình ảnh màu, mỗi pixel thường có ba giá trị (3 kênh màu: B, G, R), trong khi hình ảnh đen trắng có một giá trị độ sáng.

### Có Thể Bạn Chưa biết ?

Pixel (viết tắt của picture element) là đơn vị cơ bản nhất của một hình ảnh số và thành phần cơ bản trong quá trình biểu diễn và xử lý hình ảnh số.

Một pixel thường bao gồm các thông tin như sau:

*   **Màu Sắc**: Trong hình ảnh màu, mỗi pixel sẽ có ba giá trị tương ứng với mức độ đỏ (Red), xanh lá cây (Green), và xanh dương (Blue). Mỗi giá trị này thường nằm trong khoảng từ 0 đến 255.
*   **Độ Sáng**: Trong hình ảnh đen trắng (grayscale), mỗi pixel chỉ chứa một giá trị độ sáng. Giá trị này thường cũng nằm trong khoảng từ 0 đến 255, trong đó 0 thường là màu đen và 255 là màu trắng.
*   **Vị Trí**: Pixel cũng chứa thông tin về vị trí của nó trong bảng vẽ hoặc hình ảnh.



2.   **Kích Thước Mảng:**

*   Kích thước của mảng NumPy phản ánh độ phân giải của hình ảnh. Ví dụ, một hình ảnh có kích thước 300x400 sẽ có một mảng NumPy kích thước (300, 400, 3) nếu là hình ảnh màu (3 kênh màu) hoặc (300, 400) nếu là hình ảnh đen trắng.

3.   **Thao Tác Với Pixel:**

*   Bạn có thể trực tiếp truy cập và thay đổi giá trị của từng pixel trong mảng NumPy. Điều này rất hữu ích khi bạn muốn thay đổi nội dung của hình ảnh.

4.   **Một Số Hàm Vẽ và Chèn Văn Bản:**

*   cv2.line(): Vẽ đường thẳng.
*   cv2.rectangle(): Vẽ hình chữ nhật.
*   cv2.circle(): Vẽ hình tròn.
*   cv2.putText(): Chèn văn bản.

5.   **Hiệu Suất Cao:**

*   Mảng NumPy trong OpenCV được tối ưu hóa để hiệu suất cao trong quá trình xử lý hình ảnh. Điều này làm cho OpenCV trở thành một thư viện mạnh mẽ cho các ứng dụng thị giác máy tính và xử lý hình ảnh trong thời gian thực.



1.   cv2.line(img, pt1, pt2, color[, thickness[, lineType[, shift]]]):

* Vẽ đường thẳng giữa hai điểm pt1 và pt2 trên hình ảnh img.
*   color: Màu của đường thẳng.
*   thickness: Độ dày của đường thẳng. Nếu là âm, đường thẳng sẽ được vẽ bằng màu đầy đủ.
*   lineType: Loại đường
*   shift: Số bit để dịch phải.

VÍ DỤ:

```python
import cv2
import numpy as np

# Tạo một hình ảnh trắng (một bảng vẽ)
img = np.zeros((300, 400, 3), dtype="uint8")

# Điểm đầu (pt1) và điểm cuối (pt2) của đường thẳng
pt1 = (50, 150)
pt2 = (300, 150)

# Màu của đường thẳng (xanh lá cây)
color = (0, 255, 0)

# Độ dày của đường thẳng
thickness = 10

# Loại đường (cv2.LINE_AA cho đường liền với đường góc được làm mịn)
lineType = cv2.LINE_8

# Vẽ đường thẳng lên hình ảnh
cv2.line(img, pt1, pt2, color, thickness, lineType)

# Hiển thị hình ảnh chứa đường thẳng
cv2.imshow("picture", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```


2.   cv2.rectangle(img, pt1, pt2, color[, thickness[, lineType[, shift]]]):


*   pt1: Tọa độ của góc trái trên
*   pt2: Tọa độ của góc phải dưới

VÍ DỤ:

```python
import cv2
import numpy as np

# Tạo hình ảnh trắng (nền)
height, width = 300, 400
img = np.ones((height, width, 3), dtype=np.uint8) * 255

# Đặc điểm của hình chữ nhật
pt1 = (50,50)  # Tọa độ của góc trái trên
pt2 = (200, 200)  # Tọa độ của góc phải dưới
color = (0, 0, 255)  # Màu của hình chữ nhật (màu đỏ)
thickness = 2  # Độ dày của đường viền (nếu là âm, hình chữ nhật sẽ được vẽ bằng màu đầy đủ)

# Vẽ hình chữ nhật
cv2.rectangle(img, pt1, pt2, color, thickness)

# Hiển thị hình ảnh
cv2.imshow("picture", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```



3.   cv2.circle(img, center, radius, color[, thickness[, lineType[, shift]]]):

*   center là tọa độ của tâm của hình tròn.
* radius là bán kính của hình tròn.
* color là màu của hình tròn.
* thickness là độ dày của đường viền của hình tròn.







```python
import cv2
import numpy as np

# Tạo hình ảnh trắng (nền)
height, width = 300, 400
img = np.ones((height, width, 3), dtype=np.uint8) * 255

# Đặc điểm của hình tròn
center = (150, 150)  # Tọa độ tâm của hình tròn
radius = 50  # Bán kính của hình tròn
color = (0, 255, 0)  # Màu của hình tròn (màu xanh lá cây)
thickness = 2  # Độ dày của đường viền (nếu là âm, hình tròn sẽ được vẽ bằng màu đầy đủ)

# Vẽ hình tròn
cv2.circle(img, center, radius, color, thickness)

# Hiển thị hình ảnh
cv2.imshow("picture", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

4. cv2.putText(img, text, org, fontFace, fontScale, color[, thickness[, lineType[shift]]])

*   img: Hình ảnh đích.
* text: Chuỗi văn bản cần chèn.
* org: Tọa độ (x, y) của điểm bắt đầu văn bản.
* fontFace: Kiểu font
* fontScale: Tỉ lệ kích thước của font.
* color: Màu của văn bản.
* thickness : Độ dày của đường viền văn bản.
* lineType : Loại đường.
* shift : Nếu là True, org được coi là góc trái dưới của văn bản. Nếu là False (mặc định), org được coi là góc trái trên của văn bản.




```python
import cv2
import numpy as np

# Tạo hình ảnh trắng (nền)
height, width = 300, 400
img = np.ones((height, width, 3), dtype=np.uint8) * 255

# Văn bản cần chèn
text_to_insert = "Hello, OpenCV!"

# Tọa độ và font
org = (50, 150)
font_face = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1
text_color = (0, 0, 255)  # Màu đỏ

# Chèn văn bản vào hình ảnh
cv2.putText(img, text_to_insert, org, font_face, font_scale, text_color, 2, cv2.LINE_AA)

# Hiển thị hình ảnh
cv2.imshow("picture", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Face detection (Bonus)

```python
image = resizeFrame(cv2.imread(r'assets\A_Tien2.jpg'), 0.25)

grey = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)

face_cascade = cv2.CascadeClassifier(r'assets/haarcascade_frontalface_default.xml')

faces = face_cascade.detectMultiScale(grey, 1.3, 5)

for (x,y,w,h) in faces:
     cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),2)

cv2.imshow('img',cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

cv2.waitKey(0)
```

```python
import cv2
face_cascade = cv2.CascadeClassifier(r'assets/haarcascade_frontalface_default.xml')

# Create a VideoCapture object to capture video from the camera
cap = cv2.VideoCapture(0)

while True:
    # Read the frame from the camera
    ret, frame = cap.read()

    grey = cv2.cvtColor(frame , cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(grey, 1.3, 5)

    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)

    cv2.imshow('img',cv2.cvtColor(frame, 1))

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the VideoCapture object and close the window
cap.release()
cv2.destroyAllWindows()
```
