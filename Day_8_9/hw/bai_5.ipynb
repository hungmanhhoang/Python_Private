# Statistic Data

```python
import pandas as pd
```

```python
sale = pd.read_csv('../data/sale.csv')
car = pd.read_csv('../data/car.csv')    
display(sale.head())
display(car.head())
```

```python
# Tính tổng số lượng bán ô tô trên thị trường
sale['Sales_in_thousands'].sum()
```

```python
# Tính giá trung bình của ô tô
sale['Price_in_thousands'].mean()
```

```python
# Xác định 10 ID có giá trị bán lại cao nhất và in ra thông tin xe từ DataFrame Car của 10 xe đó.
sale.sort_values(by='Price_in_thousands', ascending=False).head(10).merge(car, on='ID')
```

```python
# Đếm số lượng phương tiện mỗi loại (Passenger, Car)
car['Vehicle_type'].value_counts()
```

```python
# Xác định mode của hãng xe
car['Manufacturer'].mode()
```

```python
# Sắp xếp DataFrame Sale giảm dần theo mức độ ưu tiên số lượng bán, giá bán, giá bán lại dự kiến và in kết quả ra màn hình.
sale.sort_values(by=['Sales_in_thousands', 'Price_in_thousands', '__year_resale_value'], ascending=False)
```

```python
# Gộp 2 DataFrame lại theo thuộc tính “ID” và in kết quả ra màn hình.

sale.merge(car, on='ID')
# pd.concat([sale, car], axis=1) # or 
# or sale.join(car, on='ID')
```
