### 6. List
#### 6.1 append()
```py
numbers = [1,2,3,4]​
numbers.append(5)​
print(numbers)​
#Output: [1, 2, 3, 4, 5]
```
#### 6.2 insert()
```python
numbers = [1,2,3,4]​
# Insert 5 into the second index (List index starts with 0)​
numbers.insert(1,5)​
print(numbers)​
#Output: [1, 5, 2, 3, 4]
```
#### 6.3 extend()
```python
list1 = [1,2,3,4]​
list2 = [8,9]​
list1.extend(list2)​
print(list1)​
#Output: [1, 2, 3, 4, 8, 9]​
```
#### 6.4 remove()
```python
numbers = [1,4,2,4,5]​
numbers.remove(4)​
print(numbers)​
#Output: [1, 2, 4, 5]
```
#### 6.5 pop()​
```python
numbers = [1,4,2,4,5]​

first_pop = numbers.pop()​

print(numbers)​

print(first_pop)​

#Output: [1, 4, 2, 4]​

#Output: 5​

second_pop = numbers.pop(0)​

print(numbers)​

print(second_pop)​

#Output: [4, 2, 4]​

#Output: 1​
```

#### 6.6 sort()
```python
numbers = [3,2,1,5,4]​

# sort by ascending order​

numbers.sort()​

print(numbers) #Output: [1, 2, 3, 4, 5]​

# sort by descending order​

numbers.sort(reverse=True)​

print(numbers) #Output: [5, 4, 3, 2, 1]
```

#### 6.7 reverse()
```python
numbers = [8,6,4,2]​

numbers.reverse()​

print(numbers)​

#Output: [2, 4, 6, 8]​
```

#### 6.8 max()​

### 7. Set​
```python
# Syntax​

# my_set = {element1, element2, ...}​
myset = {1,2,4,4,6}​
print(myset)​

#Output: {1, 2, 4, 6}
```

#### 7.1 Add an element
```python
myset = {1,2,3,4,5}
myset.add(6)
print(myset)
#output: {1,2,3,4,5,6}
```

#### 7.2 Remove an element
```python
myset = {1,2,3,4,5}
myset.remove(5)
print(myset)
#output: {1, 2, 3, 4}
myset.discard(8)
print(myset)
# 8 now is not in the set
# using discard won't be fail
# if we try myset.remove(8) will be failed
#output: {1, 2, 3, 4}
```

#### 7.3 union() or |​
```python
set1 = {1, 2, 3}​
set2 = {3, 4, 5}​
set3 = set1.union(set2)​
# set3 is {1, 2, 3, 4, 5}​
set4 = set1 | set2​
# set4 is {1, 2, 3, 4, 5}
```

#### 7.4 intersection() or &
```python
set1 = {1, 2, 3}​
set2 = {3, 4, 5}​
set3 = set1.intersection(set2)​
# set3 is {3}​
set4 = set1 & set2​
# set4 is {3}​
```

#### 7.5 difference() or - 
```python
set1 = {1, 2, 3}​
set2 = {3, 4, 5}​
set3 = set1.difference(set2)​
# set3 is {1, 2}​
set4 = set1 - set2​
# set4 is {1, 2}​
```

#### 7.6 symmetric_difference() or ^
```python
set1 = {1, 2, 3}​
set2 = {3, 4, 5}​
set3 = set1.symmetric_difference(set2)​
# set3 is {1, 2, 4, 5}​
set4 = set1 ^ set2​
# set4 is {1, 2, 4, 5}​
```

#### 7.7 issubset() and issuperset()
```python
set1 = {1, 2, 3}​
set2 = {2, 3}​
set3 = {1, 2, 3, 4}​
set2.issubset(set1) # return True​
set3.issuperset(set1) # return True​
```
